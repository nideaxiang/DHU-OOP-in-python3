## 5 总结
p	打印表达式的值。
pp	漂亮地打印表达式的值。
n	继续执行，直到到达当前函数中的下一行或返回。
s	执行当前行并在第一个可能的情况下停止（在调用的函数中或在当前函数中）。
c	继续执行，仅在遇到断点时停止。
unt	继续执行，直到到达数字大于当前数字的行。使用行号参数，继续执行，直到到达数字大于或等于该值的行。
l	列出当前文件的源代码。如果没有参数，则在当前行周围列出 11 行或继续上一个列表。
ll	列出当前函数或帧的整个源代码。
b	没有参数，列出所有中断。使用行号参数，在当前文件中的此行处设置断点。
w	打印堆栈跟踪，底部显示最新的帧。箭头指示当前帧，它决定了大多数命令的上下文。
u	将堆栈跟踪中的当前帧计数（默认帧数）级别上移（移动到较旧的帧）。
d	将堆栈跟踪中的当前帧计数（默认帧数）级别向下移动（到较新的帧）。
h	请参阅可用命令的列表。
h <topic>	显示命令或主题的帮助。
h pdb	显示完整的 pdb 文档。
q	退出调试器并退出。
